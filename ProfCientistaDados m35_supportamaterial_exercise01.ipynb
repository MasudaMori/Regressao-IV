Este modelo deve prever a varivel ```renda``` com base nas demais variveis exceto ```data_ref``` e ```index```. J vimos que a varivel renda  melhor modelada com a transformao ```log()```, no se esquea disso.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from scipy.stats import shapiro
import scipy.stats as ss
from statsmodels.nonparametric.smoothers_lowess import lowess
from scipy.interpolate import interp1d
````

````python
# Lendo o arquivo CSV com os dados e removendo a coluna 'index' que no  necessria
df = pd.read_csv(filepath_or_buffer="previsao_de_renda_II.csv").drop(columns="index")

# Exibindo informaes sobre o DataFrame, como tipos de dados e quantidade de valores nulos
df.info()

# Mostrando o contedo do DataFrame
df
````

````python
def amostra(data: pd.DataFrame, column: str, perc: float = 1, seed: int = None) -> pd.DataFrame:

    # Verifica se a porcentagem est entre 0 e 1. Caso contrrio, gera um erro
    if not (0 < perc <= 1):
        raise ValueError("O parmetro perc deve estar entre 0 e 1.")

    # Verifica se a coluna especificada existe no DataFrame. Caso no exista, gera um erro
    if column not in data.columns:
        raise KeyError(f"A coluna '{column}' no existe no DataFrame.")

    # Realiza a amostragem dos dados agrupados pela coluna especificada
    # O 'frac=perc' define a frao de dados a ser amostrada, e 'random_state=seed' garante a reprodutibilidade
    new_df = (data.set_index(keys=column)
                  .groupby(by=column, group_keys=False)
                  .apply(lambda x: x.sample(frac=perc, random_state=seed), 
                         include_groups=False)
                  .reset_index())
    
    # Exibe o tamanho dos DataFrames antes e depois da amostragem
    print(f"Original DataFrame size:\n{data.groupby(by=column).size()}\nTotal para 100% dos dados: {len(data)} linhas\n")
    print(f"New DataFrame size:\n{new_df.groupby(by=column).size()}\nTotal para {perc:.1%} dos dados: {len(new_df)} linhas\n")
    
    return new_df
````

````python
# Criando uma cpia do DataFrame original para preservar os dados originais
df_ = df.copy()

# Criando uma nova coluna 'renda_log' com a transformao logartmica da coluna 'renda'
df_.loc[:, "renda_log"] = np.log(df_.renda)

# Chamando a funo 'amostra' para pegar 1% dos dados, agrupando pela coluna 'data_ref' e utilizando uma semente para reprodutibilidade
df_ = amostra(data=df_, column="data_ref", perc=.01, seed=420)

# Exibindo o DataFrame resultante
df_
````

````python
# Preenchendo os valores nulos da coluna 'tempo_emprego' com a mdia dos valores dessa mesma coluna
df_.fillna(value={"tempo_emprego": df_.tempo_emprego.mean()}, inplace=True)

# Exibindo a contagem de valores nulos em cada coluna aps o preenchimento
df_.isna().sum()
````

````python
# Dividindo o DataFrame em conjuntos de treino e teste, com 80% para treino e 20% para teste
# A semente (random_state=420) garante que a diviso seja reprodutvel
df_train, df_test = train_test_split(df_, test_size=.2, random_state=420)

# Exibindo as dimenses (nmero de linhas e colunas) dos DataFrames de treino e teste
print(df_train.shape)  # Mostra o tamanho do conjunto de treino
print(df_test.shape)   # Mostra o tamanho do conjunto de teste
````

````python
# Criando e ajustando um modelo de regresso linear usando a frmula 'renda_log ~ tempo_emprego'
# O modelo  treinado com o DataFrame de treino (df_train)
# O mtodo .fit() ajusta o modelo aos dados e .summary() exibe um resumo dos resultados do modelo
smf.ols(formula="renda_log ~ tempo_emprego", data=df_train).fit().summary()
````

````python
# Aplicando a suavizao LOWESS (Locally Weighted Scatterplot Smoothing) para ajustar uma curva aos dados
# 'endog'  a varivel dependente (renda_log) e 'exog'  a varivel independente (tempo_emprego)
# A funo sm.nonparametric.lowess realiza o ajuste e armazena o resultado na varivel 'lowess'
lowess = sm.nonparametric.lowess(endog=df_train.renda_log, exog=df_train.tempo_emprego)
````

````python
# Criando uma funo de interpolao 1D usando os dados ajustados pelo LOWESS
# 'x' so os valores da varivel independente e 'y' so os valores da varivel dependente ajustados
# 'bounds_error=False' permite extrapolao fora do intervalo dos dados fornecidos
f_ = interp1d(x=lowess[:, 0], y=lowess[:, 1], bounds_error=False)
````

````python
# Aplicando a funo de interpolao LOWESS aos dados de treino e teste
# Adicionando a coluna 'lowess' aos DataFrames de treino e teste com os valores ajustados pela funo de interpolao
df_train["lowess"] = f_(df_train.tempo_emprego)
df_test["lowess"] = f_(df_test.tempo_emprego)
````

````python
# Criando e ajustando um modelo de regresso linear onde 'renda_log'  a varivel dependente e 'lowess'  a varivel independente
# Exibindo o resumo do modelo ajustado para avaliar os resultados
reg_lowess = smf.ols(formula="renda_log ~ lowess", data=df_train).fit()
reg_lowess.summary()
````

````python

%%time

# Obtendo as cores padro para os grficos a partir das configuraes do Matplotlib
colors = plt.rcParams["axes.prop_cycle"].by_key()["color"]

# Criando um grfico de disperso para os dados de treino
plt.scatter(x=df_train.tempo_emprego, y=df_train.renda_log, color=colors[0], alpha=.5)
plt.scatter([], [], color=colors[0], label="Dados de Treino")  # Adicionando uma entrada na legenda para os dados de treino

# Criando um grfico de disperso para os dados de teste
plt.scatter(x=df_test.tempo_emprego, y=df_test.renda_log, color=colors[1], alpha=.5)
plt.scatter([], [], color=colors[1], label="Dados de Teste")  # Adicionando uma entrada na legenda para os dados de teste

# Adicionando a curva de regresso estimada para os dados de teste
plt.plot(df_test.tempo_emprego, reg_lowess.predict(df_test), "|-", color=colors[3], label="Curva de Regresso Estimada")

# Adicionando rtulos aos eixos
plt.xlabel("tempo_emprego")
plt.ylabel("renda_log")

# Exibindo a legenda e o grfico
plt.legend()
plt.show()
````

````python
# Calculando o coeficiente de determinao (R) para avaliar a qualidade do ajuste do modelo
# O R indica a proporo da variabilidade dos dados que  explicada pelo modelo
# Utilizando o mtodo r2_score para comparar os valores reais (y_true) com as previses do modelo (y_pred)
print("{:.2%}".format(r2_score(y_true=df_test.renda_log, y_pred=reg_lowess.predict(df_test))))
````

````python
# Realizando uma regresso linear mltipla
# Varivel dependente: 'renda_log'
# Variveis independentes: 'lowess', 'tipo_renda' e 'posse_de_imovel'
reg_multi = smf.ols(formula="renda_log ~ lowess + tipo_renda + posse_de_imovel", data=df_train).fit()

# Exibindo o resumo dos resultados da regresso
reg_multi.summary()
````

````python
# Realizando um teste de identificao de outliers no modelo ajustado
# A funo outlier_test() avalia a influncia dos pontos nos resultados do modelo, incluindo resduos estudantizados e p-valores associados
outlier_test = reg_multi.outlier_test()
````

````python
# Exibindo a contagem de observaes para cada combinao de 'tipo_renda' e 'posse_de_imovel' no DataFrame de treino
# Isso ajuda a entender a distribuio das observaes por essas categorias
display(df_train.groupby(["tipo_renda", "posse_de_imovel"]).size().to_frame(name="qtd"))

# Calculando a alavancagem (leverage) para cada observao com base no modelo de regresso mltipla
# A alavancagem mede o impacto potencial de cada observao na estimativa dos parmetros do modelo
df_train["leverage"] = reg_multi.get_influence().hat_matrix_diag

# Exibindo as observaes com alavancagem maior que 0.2
# Essas observaes tm um impacto potencialmente alto na estimao dos parmetros do modelo
display(df_train[df_train.leverage > .2])

# Exibindo a contagem de observaes com alta alavancagem para cada combinao de 'tipo_renda' e 'posse_de_imovel'
# Isso fornece uma viso detalhada da distribuio das observaes com alta alavancagem por categoria
display(df_train[df_train.leverage > .2].groupby(by=["tipo_renda", "posse_de_imovel"]).size().to_frame("qtd"))
````

````python
# Filtrando os dados para remover as observaes onde 'tipo_renda'  igual a 'Bolsista'
# Cria cpias dos DataFrames de treino e teste sem as observaes de 'Bolsista'
df_train = df_train[df_train["tipo_renda"] != "Bolsista"].copy()
df_test = df_test[df_test["tipo_renda"] != "Bolsista"].copy()

# Calculando os resduos estudantizados para o DataFrame de treino com base no modelo 'reg_multi'
df_train["student_resid"] = outlier_test.student_resid

# Filtrando os dados para remover observaes com resduos estudantizados fora do intervalo [-2, 2]
# Isso ajuda a eliminar possveis outliers
df_train = df_train[(df_train["student_resid"] < 2) & (df_train["student_resid"] > -2)]

# Exibindo as dimenses dos DataFrames de treino e teste aps a filtragem
print(df_train.shape)  # Tamanho do DataFrame de treino
print(df_test.shape)   # Tamanho do DataFrame de teste
````

## Ajustes

Faa os ajustes que julgar necessrios no modelo e compare as mtricas de desempenho do modelo original e ajustado na base de testes.

````python
# Criando um grfico de linha para mostrar a mdia de 'renda_log' ao longo do tempo (data_ref)

plt.figure(figsize=(16,6))  # Define o tamanho da figura

# Agrupa os dados de treino por 'data_ref' e calcula a mdia de 'renda_log' para cada data
sns.lineplot(data=df_train.groupby(by="data_ref").renda_log.mean())

# Rotaciona os rtulos do eixo x para que as datas fiquem mais legveis
plt.xticks(rotation=45)

# Exibe o grfico
plt.show()
````

````python
# Ajustando 'renda_log' para remover a influncia da varivel 'data_ref'
# Subtrai a mdia de 'renda_log' para cada data de referncia ('data_ref') de cada valor individual de 'renda_log'
# O objetivo  ajustar a varivel 'renda_log' sem a influncia de 'data_ref', conforme avaliado no grfico anterior

df_train["renda_log_data_ref"] = df_train.renda_log - df_train.data_ref.map(df_train["renda_log"].groupby(by=df_train.data_ref).mean())

# Exibindo o DataFrame com a nova coluna 'renda_log_data_ref' que representa 'renda_log' ajustada
df_train
````

````python
# Realizando uma nova regresso linear mltipla ajustada
# A varivel dependente  'renda_log_data_ref', que representa 'renda_log' ajustada para remover o efeito da 'data_ref'
# As variveis independentes so: 'tempo_emprego', 'C(tipo_renda)' (varivel categrica codificada como dummy), e 'C(posse_de_imovel)' (outra varivel categrica codificada como dummy)

reg_ajuste = smf.ols(formula="renda_log_data_ref ~ tempo_emprego + C(tipo_renda) + C(posse_de_imovel)", data=df_train).fit()

# Exibindo o resumo dos resultados da nova regresso ajustada
reg_ajuste.summary()
````

````python
# Calcula a mdia de 'renda_log' para cada 'data_ref' no DataFrame de treino
data_map = df_train.renda_log.groupby(df_train.data_ref).mean()

# Estima os valores de 'renda_log' para o DataFrame de teste
# Adiciona a previso da regresso ajustada  mdia de 'renda_log' por 'data_ref'
renda_log_pred = df_test.data_ref.map(data_map) + reg_ajuste.predict(df_test)

# Calcula o coeficiente de determinao (R) para avaliar o desempenho do modelo ajustado
r2 = r2_score(y_true=df_test.renda_log, y_pred=renda_log_pred)

# Exibe o R como uma porcentagem
print("{:.2%}".format(r2))
````
